# Gráficos com Matplotlib

Matplotlib é uma biblioteca de software para criação de gráficos e visualizações de dados em geral, feita para e da linguagem de programação Python e sua extensão de matemática NumPy.

Links úteis:
* https://www.w3schools.com/python/matplotlib_intro.asp
* https://www.python-graph-gallery.com/
* https://realpython.com/python-matplotlib-guide/

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ano = [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000]
taxa_desemprego = [9.8, 10.1, 15.2, 3.1, 5.6, 6.7, 10.7, 20.8, 10.5]
```

```python
plt.plot(ano, taxa_desemprego)
plt.title('Taxa de desemprego vs Ano')
plt.xlabel('ano')
plt.ylabel('taxa de desemprego')
plt.show()
```

```python
x = np.linspace(-1, 1, 50)
y1 = 2*x + 1
y2 = x + 1

plt.figure(num=3, figsize=(8, 5))
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=3.0, linestyle='--')
plt.show()
```

# Gráfico de Barras

```python
data = {'C++':20, 'Python':50, 'R':40, 'MATLAB':10, 'Julia':15}

courses = list(data.keys())
estudantes = list(data.values())
```

```python
print(courses)
print(estudantes)
```

```python
plt.bar(courses, estudantes, color='maroon', width=0.5)

plt.title('Estudantes inscritos nos cursos de Linguagem de Programação')
plt.xlabel('Linguagens de Programação')
plt.ylabel('Quantidade de Alunos')

plt.show()
```

# Histogramas
É um tipo de gráfico de barras que demonstra uma distribuição de frequências. No histograma, a base de cada uma das barras representa uma classe e a altura representa a quantidade ou frequência absoluta com que o valor de cada classe ocorre.

```python
x = np.random.normal(150, 5, 300)

plt.hist(x)
plt.show()
```

```python
np.random.seed(10**7)
n_bins=20

x = np.random.randn(1000, 3)

color = ['green', 'blue', 'red']

plt.hist(x, n_bins, histtype='bar', color=color)
plt.legend(color)

plt.title('Histograma')
plt.show()
```

# Gráfico de Dispersão (ScatterPlot)

O gráfico de dispersão é um gráfico bi-dimensional para visualização de dados que usa pontos para representar o valor entre duas variáveis diferentes.

Esse tipo de gráfico é ideal quando você quer mostrar a relação entre duas variáveis para obter indícios de correlação

```python
x = np.random.normal(5.0, 1.0, 1000)
y = np.random.normal(10.0, 2.0, 1000)

plt.scatter(x, y)
plt.show()
```

```python
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

# Boxplots
Um gráfico de caixa (boxplot) é criado para exibir o resumo do conjunto de valores de dados com propriedades como mínimo, primeiro quartil, mediana, terceiro quartil e máximo.

```python
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize=(10,7))

plt.boxplot(data)

plt.show()
```
